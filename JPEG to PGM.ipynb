```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = input("Enter the name of jpeg file: ")
```

```python
p = cv2.imread(f"{filename}.jpeg", 0)

plt.imshow(p, cmap='gray')
plt.show
```

```python
rows, cols = p.shape
```

```python
with open(f"{filename}.pgm", "a+") as f:
    f.write(f"P2\n# cv.pgm\n{cols} {rows}\n255\n")
    for i in range(rows):
        for j in range(cols):
            if j != (cols-1):
                f.write(f"{p[i][j]} ")
            else:
                f.write(f"{p[i][j]}\n")
f.close()
```
